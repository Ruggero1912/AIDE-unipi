#  <font color='green'>solution 1</font>

## with matplotlib

- check the documentation of the `plt.scatter()` method

```python
plt.scatter?
```

- import the dataframe from excel

```python
composers_df = pd.read_excel('dataset/composers.xlsx', index_col = 'composer', sheet_name='Sheet5')
```

- basic solution: no grouping, no color, no legend

```python
plt.scatter(composers_df.birth,composers_df.death)
plt.xlabel('birth')
plt.ylabel('death')
```

```python
 composers_df.period
```

- intermediate solution: without legend
    - map unique periods into colors
    - define the color of each sample

```python
list(set(composers_df.period))
```

```python
my_dict = dict(zip(list(set(composers_df.period)),'rgbcyk'))
my_dict
```

```python
colors = [my_dict[x] for x in composers_df.period]
```

```python
plt.scatter(composers_df.birth,composers_df.death, color = colors)
plt.xlabel('birth')
plt.ylabel('death')
# legend missing
```

- complete solution: with legend

```python
color_map = dict(zip(composers_df.period.unique(),'rgbykc'))
plt.scatter(composers_df.birth, composers_df.death, color = composers_df.period.apply(lambda x:color_map[x]))
for k,v in color_map.items():
    plt.scatter(composers_df[composers_df.period == k].birth, 
                composers_df[composers_df.period == k].death, 
                color = v, 
                label = k)
plt.legend()
plt.xlabel('birth')
plt.ylabel('death')
```

## with `pd.DataFrame.groupby` & `pd.DataFrame.plot`

```python
fig, ax = plt.subplots(figsize=(6, 6))
grouped = composers_df.groupby('period')
for key, group in grouped:
    group.plot(ax = ax, kind='scatter', x='birth', y='death', label=key, color=color_map[key])
plt.show()
```

## with seaborn

```python
sns.scatterplot(x='birth', y='death', data=composers_df, hue='period')
```

```python
df = pd.read_csv(os.path.join('dataset','healthCovid19.csv'),index_col = 0)
df.head(10)
```

Let's add domain information: create a new column with the continent

```python
!pip install pycountry_convert
```

Note that it will raise an error if the country name is not found in the library look-up table


```python
import pycountry_convert as pc

def to_code(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")
    except KeyError:
        country_code = None
    return country_code

def to_continent(code):
    try:
        continent_name = pc.country_alpha2_to_continent_code(code)
    except KeyError:
        continent_name = None
    return continent_name

```

```python
df['Country_Code'] = df.index.to_series().apply(lambda x: to_code(x))
df['Continent'] = df['Country_Code'].apply(lambda x: to_continent(x))
```

```python
df
```

```python
df.isna().sum(axis = 0)
```

The code column can be safely deleted

```python
df = df.drop(columns = 'Country_Code',axis = 1)
```

```python
df.loc["Cote d'Ivoire","Continent"] = "AF"
```

Now we can look at aggregated information, grouping countries by continent.

```python
aggregated_df = df.groupby('Continent').mean()
```

```python
aggregated_df['Population'].plot(kind = 'bar')
plt.title('Population')
```

Which continents have the highest ratio reported for Covid19 deaths?

```python
aggregated_df['Deaths'].plot(kind = 'bar')
plt.title('Death percentage')
```

Which is the state with the highest rate of death?

```python
df.loc[df['Deaths'].idxmax()]
```

```python
df.sort_values(by='Deaths', ascending=False)
```

```python
df.sort_values(by='Deaths', ascending=False).head(8).T
```

Compare the distribution of Deaths rates among European and North American countries.

```python
deathNA = df.groupby('Continent').get_group('NA').Deaths
deathEU = df.groupby('Continent').get_group('EU').Deaths
```

```python
plt.hist(deathNA, 10, alpha=0.5, label='NA')
plt.hist(deathEU, 10, alpha=0.5, label='EU')
plt.legend(loc='upper right')
plt.xlabel('Death percentage')
plt.ylabel('# countries')
plt.show()
```
